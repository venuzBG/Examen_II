```python
import numpy as np
from scipy import optimize

# Definición del polinomio
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

# Búsqueda de raíces usando distintos intervalos
intervalos = [(-10, -5), (-5, -2), (-2, -0.5), (-0.5, 0.5), (0.5, 5)]
raices = []

for a, b in intervalos:
    try:
        r = optimize.bisect(f, a, b)
        if not any(np.isclose(r, x, atol=1e-6) for x in raices):
            raices.append(r)
    except ValueError:
        pass

# Ordenar raíces
raices = sorted(raices)

# Rellenar con NaN si faltan raíces
while len(raices) < 5:
    raices.append(np.nan)

for i, r in enumerate(raices, 1):
    print(f"x{i} =", r)
```
