# Ejercicio 1

```python
%autoreload 2
import numpy as np
from src import gauss_jacobi, gauss_seidel
import matplotlib.pyplot as plt

# Ejercicio 1
A = np.array([[2,1],[3,2]], dtype=float)
b = np.array([5,8], dtype=float)
x0 = np.array([0, 0], dtype=float)

# Resolver con Gauss-Jacobi y capturar iteraciones
def gauss_jacobi_with_history(A, b, x, tol=1e-5, max_iter=100):
    """Gauss-Jacobi que retorna la trayectoria"""
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    x = np.array(x, dtype=float, ndmin=2).T if not isinstance(x, np.ndarray) else x.reshape(-1, 1) if isinstance(x, np.ndarray) and x.ndim == 1 else x
    
    n = A.shape[0]
    history = [x.flatten().copy()]
    
    for k in range(1, max_iter + 1):
        x_new = np.zeros((n, 1))
        for i in range(n):
            suma = sum([A[i, j] * x[j] for j in range(n) if j != i])
            x_new[i] = (b[i] - suma) / A[i, i]
        
        history.append(x_new.flatten().copy())
        
        if np.linalg.norm(x_new - x) < tol:
            return np.array(history)
        x = x_new.copy()
    
    return np.array(history)

# Resolver ejercicio 1
trajectory_ej1 = gauss_jacobi_with_history(A, b, x0, tol=1e-5, max_iter=100)
print("Ejercicio 1 - Trayectoria:")
print(trajectory_ej1)
print(f"Solución final: {trajectory_ej1[-1]}")
```
